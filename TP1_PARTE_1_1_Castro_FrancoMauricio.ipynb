# ***Parte 1a: ***
Implementar el algoritmo de pasaje a coordenadas cromáticas

Se importan las librerías:

```python
%matplotlib inline  
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

Se carga el material del repositorio

```python
cd vision_computadora_I/Material_TPs/TP1/coord_cromaticas
```

Se cargan las imágenes sin transformación

```python
img1 = cv.imread('CoordCrom_1.png')
img2 = cv.imread('CoordCrom_2.png')
img3 = cv.imread('CoordCrom_3.png')

plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
```

Se convierten las imágenes originales a RGB

```python
imgRGB1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
imgRGB2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
imgRGB3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)

plt.imshow(imgRGB1)
plt.show()
plt.imshow(imgRGB2)
plt.show()
plt.imshow(imgRGB3)
plt.show()
```

Se desamblan los canales (RGB) de cada imagen para aplicar alguna de las técnicas de transformación por coordenadas cromáticas (sumar sesgo a cada canal o para hacer balance de color)

```python
im1_r, im1_g, im1_b = cv.split(imgRGB1)
im2_r, im2_g, im2_b = cv.split(imgRGB2)
im3_r, im3_g, im3_b = cv.split(imgRGB3)
```

Se hace un ajuste de balance de color utilizando el producto de la división cada canal por la sumatoria de todos los canales (R+G+B)


De este modo la imagen generada es invariante a los cambios de contraste

```python
imgRGB12 = cv.merge((im1_r/(im1_r+im1_g+im1_b), im1_g/(im1_r+im1_g+im1_b), im1_b/(im1_r+im1_g+im1_b)))
imgRGB22 = cv.merge((im2_r/(im2_r+im2_g+im2_b), im2_g/(im2_r+im2_g+im2_b), im2_b/(im2_r+im2_g+im2_b)))
imgRGB32 = cv.merge((im3_r/(im3_r+im3_g+im3_b), im3_g/(im3_r+im3_g+im3_b), im3_b/(im3_r+im3_g+im3_b)))

plt.imshow(imgRGB12)
plt.show()
plt.imshow(imgRGB22)
plt.show()
plt.imshow(imgRGB32)
plt.show()
```

Complemento: Se muestra la transformación de la imagen de color sumando un bias a cada canal (RGB); 
Las imágenes de salida podrian ser variante a los cambios de contraste.

```python
imgRGB12a = cv.merge((im1_r*4, im1_g*1, im1_b*0))
imgRGB22a = cv.merge((im2_r*4, im2_g*1, im2_b*0))
imgRGB32a = cv.merge((im3_r*4, im3_g*1, im3_b*0))

plt.imshow(imgRGB12a)
plt.show()
plt.imshow(imgRGB22a)
plt.show()
plt.imshow(imgRGB32a)
plt.show()
```
