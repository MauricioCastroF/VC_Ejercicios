```python
%matplotlib inline  
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
cd vision_computadora_I/Material_TPs/TP1/white_patch
```

```python
img1 = cv.imread('test_blue.png')
img2 = cv.imread('test_green.png')
img3 = cv.imread('test_red.png')
img4 = cv.imread('wp_blue.jpg')
img5 = cv.imread('wp_green.png')
img6 = cv.imread('wp_green2.jpg')
img7 = cv.imread('wp_red.png')
img8 = cv.imread('wp_red2.jpg')
```

```python
imgRGB1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
imgRGB2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
imgRGB3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
imgRGB4 = cv.cvtColor(img4, cv.COLOR_BGR2RGB)
imgRGB5 = cv.cvtColor(img5, cv.COLOR_BGR2RGB)
imgRGB6 = cv.cvtColor(img6, cv.COLOR_BGR2RGB)
imgRGB7 = cv.cvtColor(img7, cv.COLOR_BGR2RGB)
imgRGB8 = cv.cvtColor(img8, cv.COLOR_BGR2RGB)
```

**Se desamblan los canales (RGB) de cada imagen para aplicar algoritmo White Patch**

```python
im1_r, im1_g, im1_b = cv.split(imgRGB1)
im2_r, im2_g, im2_b = cv.split(imgRGB2)
im3_r, im3_g, im3_b = cv.split(imgRGB3)
im4_r, im4_g, im4_b = cv.split(imgRGB4)
im5_r, im5_g, im5_b = cv.split(imgRGB5)
im6_r, im6_g, im6_b = cv.split(imgRGB6)
im7_r, im7_g, im7_b = cv.split(imgRGB7)
im8_r, im8_g, im8_b = cv.split(imgRGB8)
```

**Se calcula la intensidad máxima de cada uno de los canales de color de las imágenes; luego se normalizan los componentes dividiendolos por su intensidad máxima**

```python
maxr1=np.amax(im1_r);maxg1=np.amax(im1_g);maxb1=np.amax(im1_b)
maxr2=np.amax(im2_r);maxg2=np.amax(im2_g);maxb2=np.amax(im2_b)
maxr3=np.amax(im3_r);maxg3=np.amax(im3_g);maxb3=np.amax(im3_b)
maxr4=np.amax(im4_r);maxg4=np.amax(im4_g);maxb4=np.amax(im4_b)
maxr5=np.amax(im5_r);maxg5=np.amax(im5_g);maxb5=np.amax(im5_b)
maxr6=np.amax(im6_r);maxg6=np.amax(im6_g);maxb6=np.amax(im6_b)
maxr7=np.amax(im7_r);maxg7=np.amax(im7_g);maxb7=np.amax(im7_b)
maxr8=np.amax(im8_r);maxg8=np.amax(im8_g);maxb8=np.amax(im8_b)

sr1=im1_r/maxr1;sg1=im1_g/maxg1; sb1=im1_b/maxb1
sr2=im2_r/maxr2;sg2=im2_g/maxg2; sb2=im2_b/maxb2
sr3=im3_r/maxr3;sg3=im3_g/maxg3; sb3=im3_b/maxb3
sr4=im4_r/maxr4;sg4=im4_g/maxg4; sb4=im4_b/maxb4
sr5=im5_r/maxr5;sg5=im5_g/maxg5; sb5=im5_b/maxb5
sr6=im6_r/maxr6;sg6=im6_g/maxg6; sb6=im6_b/maxb6
sr7=im7_r/maxr7;sg7=im7_g/maxg7; sb7=im7_b/maxb7
sr8=im8_r/maxr8;sg8=im8_g/maxg8; sb8=im8_b/maxb8
```

**Se combinan los componentes de color normalizados en una nueva imagen**

```python
wp1 = cv.merge((sr1, sg1, sb1))
wp2 = cv.merge((sr2, sg2, sb2))
wp3 = cv.merge((sr3, sg3, sb3))
wp4 = cv.merge((sr4, sg4, sb4))
wp5 = cv.merge((sr5, sg5, sb5))
wp6 = cv.merge((sr6, sg6, sb6))
wp7 = cv.merge((sr7, sg7, sb7))
wp8 = cv.merge((sr8, sg8, sb8))

plt.imshow(wp1)
plt.show()

plt.imshow(wp2)
plt.show()

plt.imshow(wp3)
plt.show()

plt.imshow(wp4)
plt.show()

plt.imshow(wp5)
plt.show()

plt.imshow(wp6)
plt.show()

plt.imshow(wp7)
plt.show()

plt.imshow(wp8)
plt.show()

```
